```python
i = 0
j = 1

phi = (5**.5 + 1)/2

for n in range(20):
    print("\tn = ", n)
    print("f(n)", (phi**n)*1.171)
    print("actual", i+j)
    i, j = j, i+j
    print("ratio", (phi**n)*1.171/ j)



# f(n)/f(n+1) ~= phi - 1
# f(n+1) = f(n) + f(n-1)
# f(n-1) = f(n-2) + f(n-3)

# r^n = r^n-1 + r^n-2
```

```python
p = 1.618
t = -.618



def f(n):
    pn = p**n
    tn = t**n
    rt5 = 5**.5
    return ((pn-tn)/rt5)*(n+1)

i = 0
j = 1
for n in range(20):
    print("\tn = ", n)
    print("f(n)", f(n))
    print("actual", i+j)
    i, j = j, i+j
    print("ratio", f(n)/ j)
```
